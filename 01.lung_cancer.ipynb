```python
# 딥러닝을 구동하는데 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리 불러오기
import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)

# 준비된 수술 환자 데이터를 불러오기
Data_set = np.loadtxt("data/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과를 X와 y로 구분하여 저장
X = Data_set[:, 0:17]
y = Data_set[:, 17]

# 딥러닝 구조를 결정(모델을 설정하고 실행)
model = Sequential()
# dense : fully connected network
model.add(Dense(30, input_dim=17, activation='relu')) #출력 30노드, 17개 input 활성함수 relu
model.add(Dense(1, activation='sigmoid')) # 이전 층의 input을 받아들이기 때문에 input_dim 자동 30
```

```python
model.summary()
```

```python
# 손실함수 정해주고, optimizer 함수 정해주는 과정 -> 모델링
# 손실을 어떻게 처리할 것이며, 어떤 방식으로 오차를 최소화 하는 방식으로 가져갈 건지 모델링 해주기
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

# 훈련시키기
model.fit(X, y, epochs=100, batch_size=10)
```
