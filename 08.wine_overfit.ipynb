```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df_pre = pd.read_csv('data/wine.csv', header=None)
df = df_pre.sample(frac=0.5)
```

```python
dataset = df.values
X = dataset[:,0:12].astype(float)
y = dataset[:,12]
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# 모델 저장 폴더 설정
import os
MODEL_DIR = './model/'
# 모델 폴더가 존재하지 않으면 만들기
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
    
modelpath="model/{epoch:02d}-{val_loss:.4f}.hdf5"

# checkpoint 객체 만들기
checkpointer = ModelCheckpoint(filepath=modelpath,\
                               monitor='val_loss', verbose=1, save_best_only=True) # 중간 과정 다 출력하기
```

- 정확도는 accuracy, 테스트셋 정확도는 val_acc, 학습셋 오차는 loss, 테스트 오차는 val_loss

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])

# 학습의 자동 중단
# 학습이 진행되어도 테스트셋 오차가 줄지 않으면 학습 멈추게 
early_stopping_callback = \
                EarlyStopping(monitor='val_loss', patience=100)

# 모델 실행
history = model.fit(X, y, epochs=3500, batch_size=500,\
         validation_split=0.33,\
                    callbacks=[early_stopping_callback])

print("\n Accuracy: %.4f" % (model.evaluate(X, y)[1]))


# 손실, 정확도 뽑아내서 그래프 그려보기 
y_vloss = history.history['val_loss'] # val_loss 뽑아서 저장
y_acc = history.history['accuracy'] # 정확도 뽑아서 저장
x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vloss, 'o', c='red', markersize=3)
plt.plot(x_len, y_acc, 'o', c='blue', markersize=3)
plt.show()
```

- 에포크가 필요 이상으로 올라가면 테스트셋 오차도 증가함  
- 과대적합
